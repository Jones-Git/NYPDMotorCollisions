```python
X = df_new.iloc[:,[9, 10, 28, 29, 30, 31, 32]]
X.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_new[["BOROUGH_BRONX", "BOROUGH_BROOKLYN", "BOROUGH_MANHATTAN", "BOROUGH_QUEENS", "BOROUGH_STATEN_ISLAND"]], df["NUMBER OF PERSONS INJURED"], test_size=0.2)
reg3 = tree.DecisionTreeRegressor(max_depth=3)
reg3 = reg3.fit(X_train, y_train)
predictions_3 = reg3.predict(X_test)
predictions_3
```

```python
((predictions_3 - y_test)**2).mean()
```

```python
reg3 = tree.DecisionTreeRegressor(max_depth = 5)
reg3 = reg3.fit(X, df_new["PERSONS_INJURED"])
tree.plot_tree(reg3)
```

# K-Nearest Neighbors

```python
k3nn = KNeighborsClassifier(n_neighbors = 5)
k3nn.fit(X_train, y_train)
y_pred = k3nn.predict(X_test)

confusion_matrix(y_test, y_pred, labels = [1,0])
```

```python
tp,fn, fp, tn = confusion_matrix(y_test, y_pred, labels = [1,0]).ravel()

sensitivity = tp/(tp + fn)
specificity = tn/(tn + fp)
precision = tp/(tp + fp)
accuracy = (tp + tn) / (tp + tn + fp + fn)

print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("Precision:", precision)
print("Accuracy:", accuracy)
```
